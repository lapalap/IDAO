```python
import pandas as pd
import numpy as np
import catboost

df = pd.read_csv("train.csv")
```

```python
df["interest"] = 1
popular = df.groupby('id3')['interest'].sum().sort_values(ascending=False).reset_index()
popular.head(10)
```

```python
popular.to_csv("popular.csv", index = False)
```

```python
df1 = df.drop_duplicates("id3")
graph1 = df1.groupby('id3')['id1'].apply(list)
graph2 = df1.groupby('id3')['id2'].apply(list)
result = pd.concat([graph1, graph2], axis=1)
dictionary = result.to_dict('index')

def get_category(id3):
    return [dictionary[id3]['id1'][0], dictionary[id3]['id2'][0]]

def get_category1(id3):
    return get_category(id3)[0]

def get_category2(id3):
    return get_category(id3)[1]

def number_of_diff_categories1(arr):
    q = []
    for i in arr:
        q.append(get_category(i)[0])
    return(len(list(set(q))))
```

```python
popular['category1'] =popular['id3'].apply(get_category1)
popular['category2'] =popular['id3'].apply(get_category2) 
popular
```

```python
popular.to_csv("popular.csv", index = False)
```
