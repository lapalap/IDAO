```python
import pandas as pd
import numpy as np
import catboost

train = pd.read_csv("train.csv")
train=train.drop(['id1','id2','id3'], axis=1)
train['interest'] = 1

train
```

```python
train = train.groupby(['user_id','date'],as_index = False).sum().pivot('user_id','date').fillna(0)
```

```python
train
```

```python
train = train.shift()[1:]
train
```

```python
test = train.loc[:, '48':'54']
train = train.loc[:, '1':'47']
```

```python
from catboost import Pool, CatBoostRegressor
```

```python
TRAIN = Pool(train, test)
model = CatBoostRegressor(iterations=500, depth=10, thread_count=12, learning_rate=0.4,border_count = 20,l2_leaf_reg=5, loss_function='RMSE')
#train the model
model.fit(TRAIN, plot = True)
```
